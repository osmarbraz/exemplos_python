## Testando

```python
import PyPDF2

# Abra o arquivo PDF em modo de leitura binária
with open('arquivo.pdf', 'rb') as pdf_file:
    # Crie um objeto PDFReader
    pdf_reader = PyPDF2.PdfReader(pdf_file)

    # Verifique se o arquivo PDF possui várias páginas
    if len(pdf_reader.pages) > 1:
        # Concatene o texto de todas as páginas
        text = ''
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            text += page.extract_text()
    else:
        # Extraia o texto da única página
        page = pdf_reader.getPage(0)
        text = page.extractText()

# Imprima o texto extraído
print(text)
```

```python
# Cria e abre o arquivo "arquivo.txt" para escrita
arquivo = open("arquivo1.txt", "w")

# Escreve no arquivo
arquivo.write(text)

# Fecha o arquivo
arquivo.close()
```

```python
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage
from io import StringIO

def convert_pdf_to_txt(path):
    rsrcmgr = PDFResourceManager()
    retstr = StringIO()
    codec = 'utf-8'
    laparams = LAParams()
    device = TextConverter(rsrcmgr, retstr, codec=codec, laparams=laparams)
    fp = open(path, 'rb')
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    password = ""
    maxpages = 0
    caching = True
    pagenos=set()

    for page in PDFPage.get_pages(fp, pagenos, maxpages=maxpages, password=password,caching=caching, check_extractable=True):
        interpreter.process_page(page)

    text = retstr.getvalue()

    fp.close()
    device.close()
    retstr.close()
    return text
```

```python
text = convert_pdf_to_txt("arquivo.pdf")
print(text)
```

```python
# Cria e abre o arquivo "arquivo.txt" para escrita
arquivo = open("arquivo2.txt", "w")

# Escreve no arquivo
arquivo.write(text)

# Fecha o arquivo
arquivo.close()
```
