## Carregando a lista de conjunto de dados disponibilizados

```python
# Import da biblioteca
from datasets import list_datasets

# Carregando a lista do conjunto de dados
datasets_list = list_datasets()

# Imprime o tamanho do conjunto de dados
print(len(datasets_list))
```

```python
for dataset in datasets_list:
    print(dataset)
```

```python
# Import da biblioteca
from datasets import load_dataset

# Carrega o conjunto de dados
datasets = load_dataset('squad_v1_pt')

print(datasets)
```

## Carregando os conjuntos de dados de treino e e validação separadamente

```python
# Import da biblioteca
from datasets import load_dataset

# Carrega o conjunto de dados
train_ds, validation_ds = load_dataset('squad_v1_pt', split=['train', 'validation'])

print(train_ds)
print(validation_ds)

#107,785₢
```
