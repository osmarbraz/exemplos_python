# Exemplo de carregamento de um conjunto de dados

## Carregando um conjunto de dados

Lista dos conjunto de dados em português.
https://huggingface.co/datasets?filter=languages:pt

```python
# Import da biblioteca
from datasets import load_dataset

# Carrega o conjunto de dados
datasets = load_dataset('squad_v1_pt')

print(datasets)
```

## Carregando os conjuntos de dados de treino e e validação separadamente

```python
# Import da biblioteca
from datasets import load_dataset

# Carrega o conjunto de dados
train_ds, validation_ds = load_dataset('squad_v1_pt', split=['train', 'validation'])

print(train_ds)
print(validation_ds)

#107,785₢
```

## Percorre o conjunto de dados

```python
for i,x in enumerate(train_ds):
  if i < 15:
    print(x)
    print(x['id'])
    print(x['title'])
    print(x['context'])
    print(x['question'])
    print(x['answers'])   
```

## exams

https://huggingface.co/datasets/exams

```python
# Import da biblioteca
from datasets import load_dataset

# Carrega o conjunto de dados
datasets = load_dataset('exams')

print(datasets)
```

Percorre o conjunto de dados

```python
cont = 0
for i, x in enumerate(datasets['validation']):    
    if x['info']['language'] == 'Portuguese':
      if cont < 10:
        print("\n", x)
        cont = cont + 1
```

## m_lama

## mkqa

https://huggingface.co/datasets/mkqa

```python
# Import da biblioteca
from datasets import load_dataset

# Carrega o conjunto de dados
datasets = load_dataset('mkqa')

print(datasets)
```

Percorre o conjunto de dados

```python
cont = 0
for i, x in enumerate(datasets['train']):    
  if i < 10:    
    print("\n",x)   
    print("\nPergunta:", x['queries']['pt'])    
    print("Resposta:", x['answers']['pt'][0]['text'])
    cont = cont + 1
```

## squad_v1_pt

https://huggingface.co/datasets/squad_v1_pt

Tradução do SQUAD 1.1 realizada automaticamente usando a API do Google Cloud.

https://rajpurkar.github.io/SQuAD-explorer/


```python
# Import da biblioteca
from datasets import load_dataset

# Carrega o conjunto de dados
datasets = load_dataset('squad_v1_pt')

print(datasets)
```

Percorre o conjunto de dados

```python
for i, x in enumerate(datasets['validation']):    
  if i < 10:    
    print("\n",x)       
    print("Pergunta:", x['question'])
    print("Contexto:",x['context'])
    print("Resposta:",x['answers'])
    for j, resposta in enumerate(x['answers']['text']):
      print("   Resposta:",resposta)
```
