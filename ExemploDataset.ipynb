## Carregando os conjuntos de dados de treino e validação separadamente

```python
# Import da biblioteca
from datasets import load_dataset

# Carrega o conjunto de dados
train_ds, validation_ds = load_dataset("squad_v2", split=["train", "validation"])

print(train_ds)
print(validation_ds)

#107,785₢
```

## Percorre o conjunto de dados

```python
for i,x in enumerate(train_ds):
  if i < 15:
    print(x)
    print(x["id"])
    print(x["title"])
    print(x["context"])
    print(x["question"])
    print(x["answers"])
```

## exams

https://huggingface.co/datasets/exams

```python
# Import da biblioteca
from datasets import load_dataset

# Carrega o conjunto de dados
dataset = load_dataset("exams")

print(dataset)
```

Percorre o conjunto de dados

```python
cont = 0
for i, x in enumerate(dataset["validation"]):
    if x["info"]["language"] == "Portuguese":
      if cont < 10:
        print("\n", x)
        cont = cont + 1
```

Apagando os arquivos

```python
dataset.cleanup_cache_files()
```

## squad_v1_pt

## bbc_news_ptbr_summary

https://huggingface.co/datasets/celsowm/bbc_news_ptbr_summary

```python
# Import da biblioteca
from datasets import load_dataset

# Carrega o conjunto de dados
dataset = load_dataset("celsowm/bbc_news_ptbr_summary")

print(dataset)
```

Percorre o conjunto de dados

```python
cont = 1
for i, x in enumerate(dataset["train"]):
      if cont < 10:
        print("\n", x)
        cont = cont + 1
```

Gerando um arquivo texto (csv) com os dados do dataset

```python
import pandas as pd

# Lista para receber os dados selecionados
dados = []

# Percorre o dataframe
for i, x in enumerate(dataset["train"]):
      # Seleciona somente 20 documentos
      if i <= 20:
        # Adiciona o id e o documento
        dados.append({
            "id": i + 1,
            "documento": x["resumo"]
        })

# Cria o DataFrame para poder salvar os dados
df = pd.DataFrame(dados)

# Salva como CSV
df.to_csv("documentos.csv", index=False, encoding="utf-8")
```

Apagando os arquivos

```python
dataset.cleanup_cache_files()
```
